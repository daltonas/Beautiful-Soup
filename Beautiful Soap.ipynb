```python
# Używając wyrażenia regularnego, pobierz z kodu źródłowego dowolnej strony domeny wikipedia.pl listę odnośników prowadzących poza tą witrynę.
# Uwaga - W kodzie witryny linki zaczynające się od znaku # oznaczają link lokalny, takie należy pominąć w wyniku.

from bs4 import BeautifulSoup
import requests
import re

www = 'https://pl.wikipedia.org/wiki/Anna'
kod_strony = requests.get(www)
sparsowany = kod_strony.text
list_stron = []
uzupelnienie_wzoru = 'https://pl.wikipedia.org'
wzor_wyrazenia = r'<a\s+href=\"([a-zA-Z\:\/\.\%0-9\_\#\(\)\-]+)\"'

for strona in re.findall(wzor_wyrazenia, sparsowany):
    if not strona.startswith('#'):
        if strona.startswith('/'):
            list_stron.append('{}{}'.format(uzupelnienie_wzoru, strona))
        else:
            list_stron.append(strona)

list_stron
```
